# Group the reason for absence

```python
df_data_copy.columns.values
```

```python
df_data_copy.drop(['Reason for Absence'], axis=1, inplace = True)
```

Here grouping is reorganising certain type of variables into groups, also known as classification. Based on the data, all the reasons can be classified into 4 groups.

```python
reason_columns.loc[:, 1:14]
```

```python
reason_type_1 = reason_columns.loc[:, 1:14].max(axis=1)
reason_type_2 = reason_columns.loc[:, 15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:, 18:21].max(axis=1)
reason_type_4 = reason_columns.loc[:, 22:28].max(axis=1)
```

```python
df_data_copy = pd.concat([df_data_copy, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis=1)
```

```python
df_data_copy
```

```python
df_data_copy.columns.values
```

# Extract the month value

```python
df_reason_mod['Date'][0]
```

```python
df_reason_mod['Date'][0].month
```

```python
list_months = []
```

# Extract the day of the week

```python
def date_to_weekday(value):
    return value.weekday()
```

```python
df_reason_mod['Day of the week'] = df_reason_mod['Date'].apply(date_to_weekday)
```

```python
df_reason_mod.drop('Date', axis = 1, inplace=True)
```
